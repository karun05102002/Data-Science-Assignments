```python
Q6. List down some best practices for exception handling in python.
```

```python
1.The statement inside the try block is executed.

2.If the statement is successful, both except clauses are skipped and the code inside the finally clause is run.

3.If the statement inside the try block fails, the code in the first except statement is executed. 
  If the statement fails due to a ValueError (i.e. not being able to convert a non-digit to an int), 
  the code in the except ValueError block is run.

4.If the statement inside the try block fails and the error is not a ValueError, the second except statement is checked.
  If the statement fails due to a ZeroDivisionError (i.e. integer is being divided by zero), the code inside
  the except ZeroDivisionError block is run.

5.The finally clause will always execute after the last task completes — regardless of whether the last task is 
  in the try block or except block
    
6.Don’t declare new variables inside a try statement that might not be reached.
```
